```python
import nltk 
import re 
import gensim
from gensim.models import Word2Vec
from nltk.corpus import stopwords
```

```python
paragraph = """Krishna is the god of love, tenderness and compassion. Hindu mythology portrays him as a prankster, a gentle lover, a universal supreme being and child-like God.(Instagram /vrindavan.krishna)
LIFESTYLE
Krishna Janmashtami 2019: The story of Lord Krishna’s birth
Krishna Janmashtami 2019: According to the tale, Krishna was born in the Yadava clan of Mathura to Queen Devaki and her husband, King Vasudeva.
Hindustan Times, Delhi | By HT Correspondent
UPDATED ON AUG 23, 2019 03:31 PM IST
Janmashtami, also known as Krishna Janmashtami or Gokulashtami, is a major Hindu festival in India and celebrates the birth of Lord Krishna, who happened to be the eighth incarnation or avatar of Lord Vishnu, the supreme God of the Hindus. The eighth day of the Bhadrapada month or Ashtami as per Hindu lunar calendar marks Janmashtami, and will be celebrated on August 24 (Saturday) this year.

It is believed that Krishna was born in a dungeon of Mathura, Uttar Pradesh, on this day at the clock struck midnight. Krishna is the god of love, tenderness and compassion. Hindu mythology portrays him as a prankster, a gentle lover, a universal supreme being and child-like God.

The narrative and stories of Lord Krishna’s life are referred to as Krishna Leela. According to the tale, Krishna was born in the Yadava clan of Mathura to Queen Devaki and her husband, King Vasudeva.
"""
```

```python
# Do Some cleaning and preprocessing :
text = re.sub(r"\[[0-9]*\]"," ",paragraph)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
sentences = nltk.sent_tokenize(text)  # Convert into sentences
```

```python
sentences
```

```python
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
sentences # Converted into word
```

```python
for i in range(len(sentences)):
    # remove stopwords
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
sentences # removed stopwords
```

```python
# Training the Word2Vec Model :
model = Word2Vec(sentences,min_count=1)
```

```python
model.wv.key_to_index  # see Count of no.
```

```python
vector = model.wv.get_vector('vishnu',norm=True) 
vector  # Vector regarding that particular words
```

```python
similar=model.wv.most_similar('love')
similar   # Similar words
```
